### Modifying the data set to include age groups and casualties

Here I will be changing the way the ages are displayed by placing them into age groups

```r
gun_violence <- gun_violence %>% mutate(Age_group = case_when(
Age <= 14 ~"0-14",
Age > 14 & Age <= 24 ~"15-24",
Age > 24 & Age <= 54 ~"25-54",
Age > 54 & Age <= 64 ~"55-64",
Age > 64 ~"65+"),

Age_group = factor(
  Age_group,
  level = c("0-14", "15-24", "25-54", "55-64", "65+")
)
)
gun_violence
```

```r
gun_violence <- gun_violence %>% mutate(casualties = `Number Killed` + `Number Injured`)
gun_violence
```

### Removing Outliers

```r
library(MASS)
```

```r
boxplot(gun_violence$casualties, main = "Casualties between 1982-2022",
ylab = "number of casualties")
```

```r
Q <- quantile(gun_violence$casualties, probs = c(.25, .75), na.rm = FALSE)
iqr <- IQR(gun_violence$casualties)
up <- Q[2]+1.5*iqr
low <- Q[1]-1.5*iqr
```

```r
gun_violence <- subset(gun_violence, gun_violence$casualties > (Q[1]-1.5*iqr) & gun_violence$casualties < (Q[2]+1.5*iqr))
dim(gun_violence)
```

```r
model_equation <- function(model, ...) {
  format_args <- list(...)

  model_coeff <- model$coefficients
  format_args$x <- abs(model$coefficients)
  model_coeff_sign <- sign(model_coeff)
  model_coeff_prefix <- case_when(model_coeff_sign == -1 ~ " - ",
                                  model_coeff_sign == 1 ~ " + ",
                                  model_coeff_sign == 0 ~ " + ")
  model_eqn <- paste(strsplit(as.character(model$call$formula), "~")[[2]], # 'y'
                     "=",
                     paste(if_else(model_coeff[1]<0, "- ", ""),
                           do.call(format, format_args)[1],
                           paste(model_coeff_prefix[-1],
                                 do.call(format, format_args)[-1],
                                 " * ",
                                 names(model_coeff[-1]),
                                 sep = "", collapse = ""),
                           sep = ""))
  return(model_eqn)
}
```

```r
model_equation(model, digits = 3, trim = TRUE)
```

```r
model_test <- lm(casualties ~ Paranoia + Location + Suicidality + Adult_Trauma + Physically_Abused
+ Bullied, data = test)
summary(model_test)
```

```r
model_equation(model_test, digits = 3, trim = TRUE)
```
